```python
import numpy as np
import pandas as pd
import random

import matplotlib.pyplot as plt
from matplotlib import ticker
```

```python
#Parámetros Generales

L = 0.1
k = 0.35
P_var = 0.25

#Parámetros 24 horas
xmin_0 = 0
xmax_0 = 24
t_0 = np.linspace(xmin_0,xmax_0,xmax_0-xmin_0 + 1)
t0_0 = 12
var = 1+np.random.uniform(-P_var,P_var,len(t_0))


d0 = (L / (1. + np.exp(-k * (t_0 - t0_0))*var))*-1
d_0 = L / (1. + np.exp(-k * (t_0 - t0_0)))*-1

#Parámetros 48 horas
xmin_1 = 0
xmax_1 = 48
t_1 = np.linspace(xmin_1,xmax_1,xmax_1-xmin_1 + 1)
t0_1 = 24
var = 1+np.random.uniform(-P_var,P_var,len(t_1))

d1 = (L / (1. + np.exp(-k * (t_1 - t0_1))*var))*-1
d_1 = L / (1. + np.exp(-k * (t_1 - t0_1)))*-1

#Tamaño del Gráfico
fig = plt.figure(figsize=(12,10))


#Curva 24 horas completa
ax = plt.subplot(2,2,1)

ax.set(title='Generación de Datos Sintéticos', ylabel = 'Deformación (mm)', xlabel = 'Tiempo (hr)')
ax.plot(t_0, d_0, 'r-', label = 'Datos de Ecuación', linewidth= 1, marker = '.', markerfacecolor='b', markeredgecolor= 'b', markersize=6 )
ax.plot(t_0, d0,'kx', label = 'Datos con Varibailidad')
plt.xlim(0,24)
plt.xticks(range(0, 25, 2))

ax.legend(loc='best')
ax.grid()


#Curva 24 horas Zoom a 12 horas
ax = plt.subplot(2,2,2)

ax.set(title='Generación de Datos Sintéticos', ylabel = 'Deformación (mm)', xlabel = 'Tiempo (hr)')
ax.plot(t_0, d_0, 'r-', label = 'Datos de Ecuación', linewidth= 1, marker = '.', markerfacecolor='b', markeredgecolor= 'b', markersize=6 )
ax.plot(t_0, d0,'kx', label = 'Datos con Varibailidad')
plt.xlim(0,13)
plt.xticks(range(0, 13, 2))

ax.legend(loc='lower left')
ax.grid()


#Curva 48 horas completa
ax = plt.subplot(2,2,3)

ax.set(title='Generación de Datos Sintéticos', ylabel = 'Deformación (mm)', xlabel = 'Tiempo (hr)')
ax.plot(t_1, d_1, 'r-', label = 'Datos de Ecuación', linewidth= 1, marker = '.', markerfacecolor='b', markeredgecolor= 'b', markersize=6 )
ax.plot(t_1, d1,'kx', label = 'Datos con Varibailidad')
plt.xlim(0,48)
plt.xticks(range(0, 49, 6))
ax.legend(loc='best')
ax.grid()


#Curva 48 horas Zoom a 12 horas
ax = plt.subplot(2,2,4)

ax.set(title='Generación de Datos Sintéticos', ylabel = 'Deformación (mm)', xlabel = 'Tiempo (hr)')
ax.plot(t_1, d_1, 'r-', label = 'Datos de Ecuación', linewidth= 1, marker = '.', markerfacecolor='b', markeredgecolor= 'b', markersize=6 )
ax.plot(t_1, d1,'kx', label = 'Datos con Varibailidad')
plt.xlim(11,25)
plt.xticks(range(12, 25, 2))
ax.legend(loc='lower left')
ax.grid()


#Generar Gráfico
fig.tight_layout(pad=4)
fig.show;

plt.savefig('Figura 1.jpeg',dpi=600)
```
